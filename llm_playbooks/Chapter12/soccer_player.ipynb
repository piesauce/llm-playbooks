```python
from transformers import pipeline
import matplotlib.pyplot as plt
import time
```

```python
# Define different model sizes
models = {
    "Small (distilgpt2)": pipeline("text-generation", model="distilgpt2"),
    "Medium (gpt2)": pipeline("text-generation", model="gpt2"),
    "Large (gpt-neo-1.3B)": pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B")
}
```

```python
# Define individuals (one lesser-known, one well-known)
individuals = {
    "Lesser Known": "Sardar Azmoun",
    "Well Known": "Cristiano Ronaldo"
}
```

```python
# Formulate questions based on Wikipedia summaries
questions = {}
for category, person in individuals.items():
    questions[category] = [
        f"When and where was {person} born?",
        f"Which teams has {person} played for?",
        f"What is {person}'s nationality?"
    ]
```

```python
# Query models and collect responses
responses = {}
for model_name, model in models.items():
    responses[model_name] = {}
    for category, qs in questions.items():
        responses[model_name][category] = []
        for q in qs:
            print(f"Querying {model_name} with: {q}")
            response = model(q, max_length=50, num_return_sequences=1)
            responses[model_name][category].append(response[0]['generated_text'])
            time.sleep(1)
```

```python
# Print responses for each model and category
for model_name, categories in responses.items():
    print(f"\nModel: {model_name}")
    for category, answers in categories.items():
        print(f"  Category: {category}")
        for i, ans in enumerate(answers):
            print(f"    Q{i+1}: {ans}")
```
